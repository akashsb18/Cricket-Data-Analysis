# Best Conversion Rate

```python
filtered_df4 = df2[(df2['100'] > 0) & (df2['50'] > 0) & (df2['100'] >= 20)].copy()

filtered_df4['ratio'] = (filtered_df4['100'] / filtered_df4['50'])*100

best_conversion=filtered_df4.sort_values(by='ratio',inplace=False,ascending=False)
best_conversion.head(20)
```
